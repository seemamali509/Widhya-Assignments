```python
#import required liabraries
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import roc_auc_score
```

```python
#read data
df=pd.read_csv("C:/Users/Seema/Desktop/DataSciencePractice/Widhya/flightdelay/flights.csv",low_memory=False)
```

```python
#take first 100000rows from the dataset
sample=df.head(100000)
```

```python
#Glancing
sample.head(5)
```

```python
#checking null values
sample.isnull().sum()
```

```python
#replacing null with mean
sample['ARRIVAL_DELAY']=sample['ARRIVAL_DELAY'].fillna(sample['ARRIVAL_DELAY'].mean())
```

```python
#replace all the column null values with their mean
sample_new=sample.fillna(sample.mean())
```

```python
sample_new.isnull().sum()
```

```python
#Microtask1: calculate number of dicerted flights
sample['DIVERTED'].sum()
```

```python
#Microtask 2 check correlation
sample.corr()
```

```python
#drop columns which doesn't have impact on the data set
sample_new1=sample.drop(['YEAR','FLIGHT_NUMBER','AIRLINE','DISTANCE','TAIL_NUMBER','TAXI_OUT','SCHEDULED_TIME','DEPARTURE_TIME','WHEELS_OFF','ELAPSED_TIME','AIR_TIME','WHEELS_ON','DAY_OF_WEEK','TAXI_IN','CANCELLATION_REASON'],axis=1,inplace=False)
```

```python
#count the numbers of flights who has delay of 15mins and count to be stored in Results column, 
#Result column acts as dependent variable
#i=1
for i in range(len(sample_new1.ARRIVAL_DELAY)):
    if sample_new1.loc[i,'ARRIVAL_DELAY']>15:
        result=1
    else:
        result=0
    sample_new1.loc[i,'Results']=result



    
```

```python
#number of flights delayed by 15mins
sample_new1['Results'].sum()
```

```python
#microtask 4
X=sample_new[['MONTH','DAY','SCHEDULED_DEPARTURE','DEPARTURE_DELAY','SCHEDULED_ARRIVAL','DIVERTED','CANCELLED','AIR_SYSTEM_DELAY','SECURITY_DELAY','AIRLINE_DELAY','LATE_AIRCRAFT_DELAY','WEATHER_DELAY']]
y=sample_new1['Results']
```

```python
#Split the data Test data as 30%
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
scaler = StandardScaler()
scaler.fit_transform(X_train,X_test)
```

```python
#Decision Tree Classifier Model
clf = DecisionTreeClassifier()
```

```python
#fit the model
clf.fit(X_train,y_train)
```

```python
#Predict y values
y_predict=clf.predict(X_test)
```

```python
#perfomrance Evaluation
roc_auc_score(y_test, y_predict)
```
